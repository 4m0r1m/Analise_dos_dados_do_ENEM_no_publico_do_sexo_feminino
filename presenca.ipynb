```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_2019 = pd.read_csv(r'C:\Users\marco\microdados\MICRODADOS_ENEM_2019.csv', sep=';', encoding='ISO-8859-1')
df_2020 = pd.read_csv(r'C:\Users\marco\microdados\MICRODADOS_ENEM_2020.csv', sep=';', encoding='ISO-8859-1')
```

## Dicionarios

```python
cor_raca_mapping = {
    0: 'Não declarado',
    1: 'Branca',
    2: 'Preta',
    3: 'Parda',
    4: 'Amarela',
    5: 'Indígena'
}

presenca_ch_mapping = {
    0: 'Faltou a prova',
    1: 'Presente na prova',
     2: 'Eliminado na prova'
}
presenca_cn_mapping = {
    0: 'Faltou a prova',
    1: 'Presente na prova',
    2: 'Eliminado na prova'
}
presenca_lc_mapping = {
    0: 'Faltou a prova',
    1: 'Presente na prova',
    2: 'Eliminado na prova'
    }
presenca_mt_mapping = {
    0: 'Faltou a prova',
    1: 'Presente na prova',
    2: 'Eliminado na prova'
}

sexo_mapping = {
    'F': 'Feminino',
    'M': 'Masculino'
}
```

## Extração dos resultados
- Presença Ciências humanas e raça/cor
- Presença Ciências da natureza e raça/cor
- Presença Linguagem e raça/cor
- Presença Matemática e raça/cor

```python

PCHSRC_2019 = df_2019.groupby(['TP_PRESENCA_CH', 'TP_SEXO', 'TP_COR_RACA']).size().reset_index(name='Variação percentual')

PCHSRC_2019['TP_COR_RACA'] = PCHSRC_2019['TP_COR_RACA'].map(cor_raca_mapping)
PCHSRC_2019['TP_PRESENCA_CH'] = PCHSRC_2019['TP_PRESENCA_CH'].map(presenca_ch_mapping)
PCHSRC_2019['TP_SEXO'] = PCHSRC_2019['TP_SEXO'].map(sexo_mapping)

print(PCHSRC_2019)

PCHSRC_2020 = df_2020.groupby(['TP_PRESENCA_CH', 'TP_SEXO', 'TP_COR_RACA']).size().reset_index(name='Variação percentual')

PCHSRC_2020['TP_COR_RACA'] = PCHSRC_2020['TP_COR_RACA'].map(cor_raca_mapping)
PCHSRC_2020['TP_PRESENCA_CH'] = PCHSRC_2020['TP_PRESENCA_CH'].map(presenca_ch_mapping)
PCHSRC_2020['TP_SEXO'] = PCHSRC_2020['TP_SEXO'].map(sexo_mapping)

print(PCHSRC_2020)

teste = (PCHSRC_2020['Variação percentual'] - PCHSRC_2019['Variação percentual']) / (PCHSRC_2019['Variação percentual']) * 100

PCHSRC_2019['Variação percentual'] = teste

PCHSRC_VARIACAO_F = PCHSRC_2019[PCHSRC_2019['TP_SEXO'] == 'Feminino']

print(PCHSRC_VARIACAO_F['TP_PRESENCA_CH'])
```

```python
sns.set(style="whitegrid")

viridis_palette = sns.color_palette("viridis", n_colors=len(PCHSRC_VARIACAO_F['TP_PRESENCA_CH'].unique()))

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='TP_COR_RACA', y='Variação percentual', hue='TP_PRESENCA_CH', data=PCHSRC_VARIACAO_F, palette=viridis_palette, ci=None)

plt.xlabel('Cor/Raça')
plt.ylabel('Variação Percentual')
plt.title('Variação Percentual por Cor/Raça e Presença na Prova de Ciências Humanas de 2019-2020')
plt.legend(title='Presença na Prova', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```
